```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
V = np.array(range(-110, 10))
exp = np.exp
a_m = 0.1*(V + 25)/(exp((V + 25)/10)-1)
b_m = 4*exp(V/18)
m_inf = a_m/(a_m + b_m)
plt.plot(V, m_inf)
plt.xlabel('V (mV)')
plt.ylabel('Rate constant (m/sec)')
plt.grid(alpha=.4,linestyle='--')
plt.title('Rate constant (m/sec) vs V')
plt.legend(['a_m', 'b_m'])
plt.show()
```
