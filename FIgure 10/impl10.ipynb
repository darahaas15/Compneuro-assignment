```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
V_h = -2
exp = np.exp
V = np.array(range(-100, 60))
h_inf = 1/(1+exp((V_h - V)/7))
```

```python
plt.xlabel('V (mV)')
plt.ylabel('h_inf')
plt.grid(alpha=.4,linestyle='--')
plt.title('h_inf vs V')

plt.plot(V,h_inf)
plt.show()
```
