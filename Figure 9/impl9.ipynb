```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
V = np.array(range(-110, 30))
exp = np.exp
a_h = 0.07 * exp(V/20)
b_h = 1/(exp((V+30)/10)+1)
```

```python
plt.xlabel('V (mV)')
plt.ylabel('Rate constant (m/sec)')
plt.grid(alpha=.4,linestyle='--')
plt.title('Rate constant (m/sec) vs V')
plt.legend(['a_h', 'b_h'])

plt.plot(V, a_h)
plt.plot(V, b_h)
plt.show()
```
